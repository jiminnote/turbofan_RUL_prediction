# 01. Data Structure Overview

이 노트북은 C-MAPSS 데이터셋 중 `train_FD001.txt` 파일의 구조를 이해하고,  
기초적인 데이터 로딩, 컬럼 정의, 결측값 확인 등을 수행합니다.

```python
# 기본 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 설정
plt.style.use('seaborn-whitegrid')
sns.set(rc={'figure.figsize': (12, 6)})
```

```python
# 컬럼명 정의 (NASA 문서 기반)
column_names = [
    'unit', 'time', 'op1', 'op2', 'op3',
    's1', 's2', 's3', 's4', 's5',
    's6', 's7', 's8', 's9', 's10',
    's11', 's12', 's13', 's14', 's15',
    's16', 's17', 's18', 's19', 's20',
    's21'
]
```

```python
df = pd.read_csv('../data/CMaps/train_FD001.txt', sep='\s+', header=None, names=column_names)
print("컬럼 개수:", df.shape[1])
```

```python

df.head()
```

```python
# 결측치 및 기본 통계
print("결측치 수:")
print(df.isnull().sum())

df.describe().T
```

```python
# std가 0인 컬럼 찾기
zero_std_cols = df.std()[df.std() == 0.0].index.tolist()
print(zero_std_cols)
```

### 엔진 수 및 각 유닛별 사이클 수 확인
→ 각 엔진이 얼마나 오래 운행되다가 고장났는지 분포 확인하기

```python
print(f"전체 엔진 수: {df['unit'].nunique()}")
df['time'].groupby(df['unit']).max().describe()
```

### 운전 조건 (op1, op2, op3) 분포 확인
→ 나중에 조건별로 데이터 스케일 차이가 크면 정규화 필요 여부 판단


```python
df[['op1', 'op2', 'op3']].describe()
df[['op1', 'op2', 'op3']].hist(bins=30, figsize=(12, 6))
plt.rcParams['font.family'] ='AppleGothic'
plt.rcParams['axes.unicode_minus'] =False
plt.suptitle("운전 조건 분포")
plt.show()
```

1.op1  
- 정규분포처럼 보이는 연속형변수  
→ 센서나 환경에 민감하게 반응할 수 있는 조건으로 보임  
- 모델 학습시 정규화가 유의미할 수 있음

2.op2
- 몇 개 구간으로 나뉜 불연속 값  
→ 정성적 조건 구간이나 내부 스위치 상태일 가능성  
→ one-hot encoding 고려 가능  

3.op3
- 완전희 고정된 값  
→ 실제 모델 학습에는 영향 없음 = 제거대상

### 3. 센서별 기본 통계 요약
→ 표준편차가 매우 작거나 고정된 센서는 제거 후보임 (ex. 전부 100.0인 센서)

```python
s_cols = [col for col in df.columns if 's' in col]
df[s_cols].describe().T[['mean', 'std', 'min', 'max']]
df[s_cols].hist(bins=30, figsize=(9, 8))
plt.suptitle("센서 값 분포")
plt.show()
```

#### 센서 값 분포 해석: 제거 대상 vs 유지 대상
값 고정 센서 :  1, 5, 6, 10, 16, 18,19  
 → 학습에 의미 있는 기여 없음. 제거 후보

분산이 낮은 센서 (노이즈 수준 약간 있음 → 판단 유보)  
- s17 : 거의 392~393 수준
- s9, s13, s14 : 왼쪽으로 치우친 분포 (비정규분포) → 로그 변환 고려 

유지 및 사용 권장 센서 (패턴 다양, 학습 가능성 높음)  
- s2, s3, s4, s7, s11, s12, s15, s20, s21 등
→ 값 변화 충분히 존재
→ RUL 예측에서 유의미한 열화 패턴을 가질 가능성 있음

### 4. 센서 간 상관관계 확인
→ 중복 정보가 많은 센서를 제거하거나 PCA 등 차원 축소 판단 근거

```python
corr = df[s_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("센서 간 상관관계 Heatmap")
plt.show()
```

1. 강한 양의 상관관계 (중복 가능성 ↑)  
- s2 ↔ s3   
- s12 ↔ s13  
- s13 ↔ s14  
- s7 ↔ s8  
→ 이들 중 하나만 선택하거나, PCA 등 차원축소 적용 고려
⸻
2. 거의 상관관계 없는 센서들 (독립적 정보)  
	•	s15, s20, s21 등 일부 센서  
→ 보존 우선, 서로 다른 신호 가능성 높음  
⸻
3. 0에 가까운 상관 (랜덤 혹은 고정값 센서들)  
	•	s1, s5, s10, s16, s18, s19  
→ 이미 분포 분석에서 제거 후보로 본 센서들이고, 상관도 낮음 → 제거 확신 강화  

### 5. 유닛 하나를 골라 센서 시계열 시각화
→ 실제로 열화 패턴이 있는지 육안 확인 가능


```python
unit1 = df[df['unit'] == 3]
unit1.plot(x='time', y=s_cols[:6], subplots=True, figsize=(15, 20), layout=(3, 2))
plt.suptitle("센서 시계열")
plt.show()  
```

제거대상 확정 : s1, s5. 
 s6 -> 일부 고정 일부 분산 : 유닛 내 variance가 0이면 모델이 시계열 특징을 못잡음

```python
# 평균 단위 분산 계산
df.groupby('unit')['s6'].std().mean()
```

## 요약
- 사용 센서: 13개 (고정값 제거 + 중복 제외)
- 운전 조건: op1, op2 유지 / op3 제거

```python
import pandas as pd
import numpy as np

# 1. 테스트셋 로드
test_df = pd.read_csv('../data/CMaps/test_FD001.txt', sep='\s+', header=None)
test_df.columns = [
    'unit', 'time', 'op1', 'op2', 'op3',
    'sensor1', 'sensor2', 'sensor3', 'sensor4', 'sensor5',
    'sensor6', 'sensor7', 'sensor8', 'sensor9', 'sensor10',
    'sensor11', 'sensor12', 'sensor13', 'sensor14', 'sensor15',
    'sensor16', 'sensor17', 'sensor18', 'sensor19', 'sensor20',
    'sensor21'
]

# 2. RUL 정답 로드
rul_df = pd.read_csv('../data/CMaps/RUL_FD001.txt', header=None, names=['RUL'])

# 3. 유닛별 마지막 row 추출
last_cycles = test_df.groupby('unit')['time'].max().reset_index()
last_cycles.columns = ['unit', 'last_time']

# 4. 실제 RUL 정답 병합
true_rul = last_cycles.copy()
true_rul['RUL'] = rul_df['RUL']

# 5. 저장
true_rul.to_csv('../models/true_rul.csv', index=False)
print("✅ true_rul.csv 저장 완료")
```
